```python
import pandas as pd
from ast import literal_eval
```

```python
exp_stats = pd.read_csv('exp_stats.csv')
```

```python
exp_stats
```

```python
backend    = 'ibmq_montreal'
cost       = 'Sato21_Nonnorm'
operator   = 'SatoLiu21'
archt      = 'linear_alternating'
shots      = 10000
num_qubits = 5
num_layers = 1
samples    = 10
row        = exp_stats.loc[   (exp_stats['backend']==backend)
                        & (exp_stats['cost']==cost)
                        & (exp_stats['operator']==operator)
                        & (exp_stats['qubits']==num_qubits)
                        & (exp_stats['layers']==num_layers)
                        #& (exp_stats['shots']==shots)
                        & (exp_stats['samples']==samples)
                        & (exp_stats['archt']==archt)
                        & (exp_stats['rotations']=="['ry']")
                        & (exp_stats['sabre']==True)
                        & (exp_stats['mapomatic']==False)
                        & (exp_stats['DD']==False)
                        & (exp_stats['mem']==False)]
```

```python
reduced = row[['backend', 'cost', 'operator', 'rhs', 'qubits', 'layers', 'parameters', 'shots', 'samples',
               'mean', 'p=50']]
reduced
```

```python
#aftermax = reduced.cnots.apply(literal_eval).apply(max)
#aftermax = pd.concat([reduced.drop('cnots',axis=1), aftermax],axis=1)
aftermax = reduced
```

```python
sort = aftermax.sort_values(by=['qubits', 'layers', 'shots'])
sort
```

```python
sort_reset = sort.reset_index(drop=True)
sort_reset
```

```python
duplicates = sort_reset.drop_duplicates(subset=['shots'], keep='last')
duplicates
```

```python
duplicates.to_csv(backend+'_'+cost+'_'+'q'+str(num_qubits)+'l'+str(num_layers)+'_shots'+'.csv')
```
