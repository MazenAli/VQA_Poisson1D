```python
target = 'ibmq_ehningen'
row    = exp_stats.loc[   (exp_stats['backend']==target)
                        & (exp_stats['relative']==True)
                        & (exp_stats['innerp']=='HTest')
                        & (exp_stats['qubits']==5)
                        & (exp_stats['qae']==False)
                        #& (exp_stats['qae_method']=='mlae')
                        #& (exp_stats['max_power']==4)
                        #& (exp_stats['new_shots']==1000)
                        & (exp_stats['squared']==False)
                        & (exp_stats['rotations']=="['ry']")
                        & (exp_stats['rhs']=="hn")
                        & (exp_stats['sabre']==True)
                        & (exp_stats['mapomatic']==False)
                        & (exp_stats['DD']==False)
                        & (exp_stats['mem']==False)]
```

```python
reduced = row[['backend', 'innerp', 'rhs', 'qubits', 'layers',
               'mean', 'qae', 'qae_method','max_power']]
```

```python
sort = reduced.sort_values(by=['qubits', 'layers'])
sort
```

```python
sort_reset = sort.reset_index(drop=True)
sort_reset
```

```python
duplicates = sort_reset.drop_duplicates(subset=['qubits', 'layers'], keep='last')
duplicates
```

```python
duplicates.to_csv(target+'_relerr_htest_q5_rxryrz_hnx.csv')
```
