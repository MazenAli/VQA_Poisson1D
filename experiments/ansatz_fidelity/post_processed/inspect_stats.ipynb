```python
import pandas as pd
```

```python
exp_stats = pd.read_csv('exp_stats.csv')
```

```python
exp_stats
```

```python
target = 'ibmq_ehningen'
row    = exp_stats.loc[(exp_stats['backend']==target)
                        & (exp_stats['archt']=='linear')
                        & (exp_stats['rotations']=="['ry']")
                        & (exp_stats['entangler']=="cz")
                        & (exp_stats['shots']==1000)
                        & (exp_stats['qubits']==5)
                        & (exp_stats['sabre']==True)
                        & (exp_stats['mapomatic']==True)
                        & (exp_stats['DD']==False)
                        & (exp_stats['mem']==False)] 
```

```python
reduced = row[['backend', 'archt', 'shots', 'rotations', 'entangler', 'sabre', 'mapomatic', 'DD', 'mem',
               'qubits', 'layers', 'parameters', 'depth', 'cnots', 'p=50',
               'p=0', 'p=5','p=25', 'p=75', 'p=95', 'p=100']]
```

```python
sort = reduced.sort_values(by=['shots', 'qubits', 'layers'])
sort
```

```python
sort_reset = sort.reset_index(drop=True)
sort_reset
```

```python
duplicates = sort_reset.drop_duplicates(subset=['qubits', 'layers'], keep='last')
duplicates
```

```python
duplicates.to_csv(target+'_q5_rycz_mapomatic.csv')
```
