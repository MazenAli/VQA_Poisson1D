```python
import pandas as pd
from ast import literal_eval
```

```python
exp_stats = pd.read_csv('exp_stats.csv')
```

```python
exp_stats
```

```python
backend    = 'ibmq_ehningen'
run_mode   = 'process_results'
operator   = 'SatoLiu21'
archt      = 'linear_alternating'
cost       = 'Sato21_Nonnorm'
#shots      = 10000
num_qubits = 5
num_layers = 3
relative   = False
samples    = 50

row        = exp_stats.loc[(exp_stats['backend']==backend)
                        & (exp_stats['run_mode']==run_mode)
                        & (exp_stats['relative']==relative)
                        & (exp_stats['operator']==operator)
                        & (exp_stats['qubits']==num_qubits)
                        & (exp_stats['layers']==num_layers)
                        & (exp_stats['cost']==cost)
                        #& (exp_stats['shots']==shots)
                        & (exp_stats['samples']==samples)
                        & (exp_stats['archt']==archt)
                        & (exp_stats['rotations']=="['ry']")
                        & (exp_stats['sabre']==True)
                        & (exp_stats['mapomatic']==False)
                        & (exp_stats['DD']==False)
                        & (exp_stats['mem']==False)]
```

```python
run_mode   = 'run_var_benchmark'
operator   = 'SatoLiu21'
archt      = 'linear_alternating'
cost       = 'Sato21_Nonnorm'
num_qubits = 5
num_layers = 3

row        = exp_stats.loc[(exp_stats['run_mode']==run_mode)
                        & (exp_stats['qubits']==num_qubits)
                        & (exp_stats['layers']==num_layers)
                        & (exp_stats['cost']==cost)
                        & (exp_stats['archt']==archt)
                        & (exp_stats['rotations']=="['ry']")]
```

```python
reduced = row[['backend', 'step_size', 'archt', 'cost', 'qubits', 'layers', 'parameters', 'shots', 'samples',
               'mean', 'p=0', 'p=5', 'p=25', 'p=50', 'p=75', 'p=95', 'p=100']]
reduced
```

```python
sort = reduced.sort_values(by=['qubits', 'layers', 'shots'])
sort
```

```python
sort_reset = sort.reset_index(drop=True)
sort_reset
```

```python
duplicates = sort_reset.drop_duplicates(subset=['shots'], keep='last')
duplicates
```

```python
duplicates.to_csv(backend+'_'+cost+'_'+'q'+str(num_qubits)+'l'+str(num_layers)+'_var'+'.csv')
```
